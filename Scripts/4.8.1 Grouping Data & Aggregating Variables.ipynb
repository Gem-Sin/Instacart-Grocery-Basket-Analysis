STEP 6. TARGETING DIFFERENT TYPES OF SPENDERS

5. Create 'spending_flag' column

```python
# Create a new column for each user's average price using transform
df_ords_prods_merge['avg_price_per_user'] = df_ords_prods_merge.groupby('user_id')['prices'].transform('mean')
```

```python
# Use an if statement to assign the spending flag
spending_flag = []

for price in df_ords_prods_merge['avg_price_per_user']:
    if price < 10:
        spending_flag.append("Low spender")
    else:
        spending_flag.append("High spender")
```

```python
# Create column 'spending_flag'
df_ords_prods_merge['spending_flag'] = spending_flag
```

```python
# Check frequency
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

STEP 7. FREQUENT VERSUS NON-FREQUENT CUSTOMERS

```python
# Create a new column with each user's median days since prior order
df_ords_prods_merge['median_days_per_user'] = df_ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform('median')
```

6. Create 'frequency_flag' column

```python
# Create the frequency flag using a for loop
order_frequency_flag = []

for days in df_ords_prods_merge['median_days_per_user']:
    if days > 20:
        order_frequency_flag.append("Non-frequent customer")
    elif days > 10:
        order_frequency_flag.append("Regular customer")
    else:
        order_frequency_flag.append("Frequent customer")
```

```python
# Add the new column to the dataframe
df_ords_prods_merge['order_frequency_flag'] = order_frequency_flag
```
