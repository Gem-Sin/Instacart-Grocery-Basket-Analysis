3. Derive New Variables

3.1 Create 'price_label' column

```python
# Run a user-defined function to create the new variable 'price_label'
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the price_label function
df['price_label'] = df.apply(price_label, axis=1)
```

4.1 Change 'busiest_day' column to 'busiest_days'

```python
# Loop for busiest 2 days and slowest 2 days
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result.append("Busiest days")
    elif value == 3 or value == 4:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")
```

```python
# Create column 
df_ords_prods_merge["busiest_day"] = result
```

The totals for the two busiest days correspond with the total rows for 0 and 1 as well as the least busy days and regularly busy.

Step 4. New Column 'busiest_period_of_day'

5. New Column 'busiest_period_of_day'

```python
# Check order frequency
df_ords_prods_merge["order_hour_of_day"].value_counts()
```

Divided the day into three periods of 8 hours. Most orders 10, 11, 14, 15, 13, 12, 16 and 9 Fewest orders 3, 4, 2, 5, 1, 0, 6, 23

```python
# Create an empty list to store labels
busiest_period_result = []
```

```python
# Loop through each hour in 'order_hour_of_day' column
for hour in df_ords_prods_merge["order_hour_of_day"]:
    if hour in [9, 10, 11, 12, 13, 14, 15, 16]:
        busiest_period_result.append("Most orders")
    elif hour in [0, 1, 2, 3, 4, 5, 6, 23]:
        busiest_period_result.append("Fewest orders")
    else:
        busiest_period_result.append("Average orders")
```

```python
# Add result list as new column
df_ords_prods_merge["busiest_period_of_day"] = busiest_period_result
```

```python
# Check frequency
df_ords_prods_merge["busiest_period_of_day"].value_counts(dropna = False)
```
