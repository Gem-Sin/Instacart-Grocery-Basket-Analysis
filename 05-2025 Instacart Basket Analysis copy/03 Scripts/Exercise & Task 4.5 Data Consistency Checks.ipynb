During my exploratory analysis of the df_prods dataframe, I noticed unusually high values in the prices column. The mean price was 9.99, but the standard deviation was over 450, which is extremely large compared to the mean. This indicated the likely presence of outliers.

To confirm this, I identified values that were more than 2 standard deviations above the mean, a common statistical method for detecting outliers. This revealed two extreme values: 14,900 and 99,999 — far outside the normal price range for products.

Since these values would heavily skew averages and distort analysis, I decided to treat them as outliers. I replaced them with the median price, which was 7.1. The median is less sensitive to extreme values than the mean, making it a more reliable measure of central tendency in this case.

After replacing these outliers, the mean price decreased to 7.6, reflecting a more accurate average for the data.
